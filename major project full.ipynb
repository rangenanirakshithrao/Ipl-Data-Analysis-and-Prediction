##  Model Creation

```python
df = pd.read_csv('matches.csv')
df
```

```python
#required columns
new_df = df[['team1', 'team2', 'winner', 'toss_decision', 'toss_winner']]
new_df.dropna(inplace=True)
```

```python
new_df.isna().sum()
```

```python
new_df
```

```python
#empty dict for replacing names with values  or converting labels into numerical values
all_teams = {}
cnt = 0
```

```python
#conversion
for i in range(len(df)):
    if df.loc[i]['team1'] not in all_teams:
        all_teams[df.loc[i]['team1']] = cnt
        cnt += 1
    if df.loc[i]['team2'] not in all_teams:
        all_teams[df.loc[i]['team2']] = cnt
        cnt += 1
```

```python
#verification of teams into values
all_teams
```

```python
#creating a new dataframe of required columns
df1=new_df.replace({"bat":1,"field":0})
```

```python
df1
```

```python
#mapping the all teams values in it
df1["team1"]=df1["team1"].map(all_teams)
df1["team2"]=df1["team2"].map(all_teams)
df1
```

```python
df1["toss_winner"]=df1["toss_winner"].map(all_teams)
df1
```

```python
df1["winner"]=df1["winner"].map(all_teams)
df1
```

```python
#taking x&y value as input and output variables
x=df1[["team1","team2","toss_winner","toss_decision"]]
x
```

```python
y=df1["winner"]
y
```

```python
#converting toss_winner column as 1&0's if team1 wins toss then it will print as 0 or else team 2 wins it's gonna print as 1
for i in range(len(x)):
    if x.iloc[i][0]==x.iloc[i][2]:
        x.iloc[i][2]=0
    else:
        x.iloc[i][2]=1
```

```python
x
```

```python
#converting winner column as 1&0's if team1 wins  then it will print as 0 or else team2 wins it's gonna print as 1
for i in range(len(x)):
    if x.iloc[i][0]==y.iloc[i]:
        y.iloc[i]=0
    else:
        y.iloc[i]=1
```

```python
y
```

```python
#the ravel function /keyword converts the 2d array as single dimension for easy prediction it's in numpy module
y=y.ravel()
```

```python
y
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x, y, test_size=0.25,random_state=1)
```

```python
#checking with decision tree
from sklearn.tree import DecisionTreeClassifier
model2 = DecisionTreeClassifier().fit(x_train, y_train)
model2.score(x_test, y_test)

```

```python
#checking with random forest
from sklearn.ensemble import RandomForestClassifier
model3 = RandomForestClassifier(n_estimators=250).fit(x, y)
model3.score(x_test, y_test)
```

```python
#checking by giving various inputs(4 values-->team1,team2,toss_winner,toss_decision)
test = np.array([2,7, 1, 0]).reshape(1,-1)
#output-->is winner
print(model3.predict(test))
```

```python
import pickle as pkl

with open('model.pkl', 'wb') as f:
    pkl.dump(model3, f)



with open('model.pkl', 'rb') as f:
    model = pkl.load(f)



model.predict(test)
```
